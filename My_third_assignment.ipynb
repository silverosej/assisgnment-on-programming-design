# Question 5: write a program that asks the user for a radius and calculates both the area and circumference of a circle(use 3.14159 for pi).

```python
radius_of_a_circle=float(input("radius of a circle"))
area_of_a_circle=((3.14159)*radius_of_a_circle**2)
circumference_of_a_circle=2*3.14159*(radius_of_a_circle)
print(f"the circumference of a circle {circumference_of_a_circle}")
print(f"the area of a circle {area_of_a_circle}")
```
