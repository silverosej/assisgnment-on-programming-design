# Question 2: write a program that determines if a year entered by the user is a leap year(hint: a leap year is divisible by 4, except for century years which must be divisible by 400).

```python
year=int(input("Enter a number: "))
if (year % 4==0 and year % 100 !=0):
     print(f"{year} is a leap year")
else:
    print(f"{year} is not a leap year")
```

## Question 5: create  a simple calculator that takes two numbers and an operator[+,-,*,/] as input and performs the corresponding operation.

```python
number1=float(input("Write a number"))
number2=float(input("Write a number"))
operator=input("Enter an operator: (+,-,*,/)")
if operator=="+":
    result=number1 + number2
elif operator== "-":
    result=number1 - number2
elif operator== "*":
    result=number1 * number2
elif operator== "/":
    if number2 != 0:
      result=number1 / number2
    else:
        result="error, division by zero is not possible"
else:
    result="invalid operator,please enter one of +,-,*,/"
print(f"the result is {result}")
```

# Question 7 create a program that determines if a triangle is equilateral(all sides equal), isosceles(two sides equal),or scalne(no sides equal) based on three side length.

```python
a=float(input("Enter the first side"))
b=float(input("Enter the second side"))
c=float(input("Enter the third side"))
if a==b==c:
   print("equilateral")
elif a==b or a==c or b==c:
    print("issoceles")
elif a!=b or b!=c or a!=c:
    print("scalene")
else:
    print("not a triangle")
```

# Question 16.create a program that determines if a word is a palindrome(reads the same forwards and backwards).

```python
word=input("Enter a word")
word_lower=word.lower()
if word_lower==word_lower[::-1]:
    print(f"{word} is a palindrome")
else:
    print(f"{word} is not a palindrome")
```

# Question 17.write a program that checks if a phone number is in the format "xxx-xxx-xxxx" where x is a digit.

```python
import re
phone_number=int(input("Enter a phone_number"))
pattern=r"^\d{3}-\d{3}-\d{4}$"
if re.match(pattern,phone_number):
    print(f"{phone_number} is correct")
else:
    print(f"{phone_number} is not correct")
```
